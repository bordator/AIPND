```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Funktion zur Erstellung der Daten
def f(x):
  return np.sin(x)

# Definition der Figureigenschaften
fig, ax = plt.subplots()
ax.set_xlim((0, 10))
ax.set_ylim((-1.5, 1.5))

# Erstellen der Linien
line, = ax.plot([], [], lw=2)

# Funktion zur Aktualisierung der Animation
def update(frame):
  # Aktualisierung der Daten
  x = np.linspace(0, 10, 100)
  y = f(x + frame/10)

  # Aktualisierung der Linie
  line.set_data(x, y)

  # Aktualisierung der Figure
  plt.draw()

# Starten der Animation
anim = FuncAnimation(fig, update, interval=10)
# Anzeigen der Animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ro')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()
```
